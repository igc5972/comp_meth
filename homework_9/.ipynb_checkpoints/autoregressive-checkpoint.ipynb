```python
import numpy as np
from matplotlib.pyplot import *
from matplotlib import rc
%matplotlib inline

# Make more readable plots
rc('font',**{'size':14})
rc('xtick',**{'labelsize':16})
rc('ytick',**{'labelsize':16})
rc('axes',**{'labelsize':18,'titlesize':18})
```

```python
def acf_plot(ax, h, rho):
    """ Make a classic ARMA-style ACF plot """
    for i in range(len(h)):
        x = h[i]
        y = rho[i]
        ax.plot([x, x], [0, y], 'k')
    ax.plot(h, rho, 'k.')
```

## Let's calculate and plot an AR(1) model

```python
np.random.seed(42)

N = 250
sigma = 1.0
phi = -0.7 #-0.7
Z_t = np.random.normal(0, sigma, N)

X_t = np.zeros(N)
X_t[0] = Z_t[0]
for i in range(1, N):
    X_t[i] = phi*X_t[i-1] + Z_t[i]
    
plot(X_t, 'k')
xlabel(r"$t$ [days]")
ylabel(r"$X_t$")
```

## The ACF of an AR(1) process

```python
h = np.arange(0, N//2)
rho = phi**h
ax = subplot(111)
acf_plot(ax, h, rho)
xlabel(r"$h$ [days]")
ylabel(r"$\rho_X(h)$")
```

## The (numerically-derived) ACF of an AR(1) process

```python
# Taken from utilities.py
lags = np.arange(-(N-1),N,dtype=np.float)
taus = np.concatenate((np.arange(1, N+1), np.arange(N-1, 0, -1)))
acf = np.correlate(X_t, X_t, "full")/taus #division by tau versus N gives a better estimate of the definition of <X_t X_t+h>
plot(lags, acf)
xlabel(r"$\tau$ [days]")
ylabel(r"ACF")
```

## The spectrum of an AR(1) process

```python
f = np.linspace(0, 0.5, 100) #0 < omega < pi ==> 0 < f < 1/2
S_X_f = sigma**2 / (1 + phi**2 - 2*phi*np.cos(2*np.pi*f))

plot(f, S_X_f)
xlabel(r"$f$ [days$^{-1}$]")
ylabel(r"$S_X(f)$")
```

## Now let's setup an AR(2) process

```python
N = 250
mode = 0

if mode == 0:
    # Eq. 3.2.20 for sunspot data in Brockwell and Davis "Introduction to Time Series and Forecasting"
    phi_1 = +1.318
    phi_2 = -0.634
    sigma = 289.2
elif mode == 1:
    # Figure 3-1
    phi_1 = 1.0/2 + 1.0/5
    phi_2 = -(1.0/2)*(1.0/5)
    sigma = 1.0
elif mode == 2:
    # Figure 3-2
    phi_1 = 9.0/10 + 1.0/2
    phi_2 = -(9.0/10)*(1.0/2)
    sigma = 1.0
elif mode == 3:
    # Figure 3-3
    phi_1 = -9.0/10 + 1.0/2
    phi_2 = -(-9.0/10)*(1.0/2)
    sigma = 1.0




Z_t = np.random.normal(0, sigma, N)

X_t = np.zeros(N)
X_t[0] = Z_t[0]
X_t[1] = Z_t[1] #not quite the right starting condition
for i in range(2, N):
    X_t[i] = phi_1*X_t[i-1] + phi_2*X_t[i-2] + Z_t[i]
    
plot(X_t, 'k.')
xlabel(r"$t$ [days]")
ylabel(r"$X_t$")
```

## The ACF of an AR(2) process

```python
h = np.arange(0, N//2)
gamma = np.zeros(len(h))
gamma[0] = ((1-phi_1)/(1+phi_2)) * sigma**2 / ((1-phi_2)**2 - phi_1**2)
gamma[1] = phi_1/(1-phi_2) * rho[0]
for i in range(2, len(h)):
    gamma[i] = phi_1*gamma[i-1] + phi_2*gamma[i-2]
    
rho = gamma/gamma[0]

ax = subplot(111)
acf_plot(ax, h, rho)
xlabel(r"$h$ [days]")
ylabel(r"$\rho_X(h)$")
```

## The spectrum of an AR(2) process

```python
f = np.linspace(0, 0.5, 100) #0 < omega < pi ==> 0 < f < 1/2
cos = np.cos(2*np.pi*f)


S_X_f = sigma**2 / (1 + phi_1**2 + 2*phi_2 + phi_2**2 + 2*phi_1*(phi_2 - 1)*cos - 4*phi_2*cos**2)

plot(f, S_X_f)
xlabel(r"$f$ [days$^{-1}$]")
ylabel(r"$S_X(f)$")
```

## Let's now look briefly at an MA(1) model

```python
np.random.seed(42)

N = 250
sigma = 1.0
theta = 0.5 #-0.5
Z_t = np.random.normal(0, sigma, N)

X_t = Z_t[1:] + theta*Z_t[:-1] 
    
plot(X_t, 'k')
xlabel(r"$t$ [days]")
ylabel(r"$X_t$")
```

## And also at an MA(2) model

```python
np.random.seed(42)

N = 250
sigma = 1.0
theta_1 = 0.3  #0.3
theta_2 = 0.5  #0.5
Z_t = np.random.normal(0, sigma, N)

X_t = Z_t[2:] + theta_1*Z_t[1:-1] + theta_2*Z_t[:-2]
    
plot(X_t, 'k')
xlabel(r"$t$ [days]")
ylabel(r"$X_t$")
```
